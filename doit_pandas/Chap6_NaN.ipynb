### 누락값 개수 구하기

```python
#누락값이 아닌 데이터 개수

ebola = pd.read_csv('./data/country_timeseries.csv')
ebola.count()
```

```python
#count() 사용하기

num_row = ebola.shape[0]
num_nan = num_row - ebola.count()
num_nan
```

```python
#count_nonzero

nonzero_all = np.count_nonzero(ebola.isnull())
nonzero_all
```

```python
#count_nonzero

nonzero_guinea = np.count_nonzero(ebola['Cases_Guinea'].isnull())
nonzero_guinea
```

```python
#value_counts()

ebola.Cases_Guinea.value_counts(dropna=False).head()
```

## 누락값 처리하기

### fillna() : 데이터프레임의 크기가 매우 크고 메모리를 효율적으로 사용해야 하는 경우에 자주 사용

```python
#누락값 0으로 대체

ebola.fillna(0)
```

```python
#method='ffill' -> 누락값의 이전값으로 대체됨
#첫번째 값이 누락값인 경우 대체 안 됨

ebola.fillna(method='ffill')
```

```python
#method='bfill' ->  누락값의 다음 값으로 변경됨
#마지막 값이 누락값인 경우 대체 안 됨

ebola.fillna(method='bfill')
```

```python
#interpolate() -> 누락값의 양쪽 값의 중앙값으로

ebola.interpolate()
```

### 누락값 삭제하기
- 무작정 삭제하면 데이터가 너무 편향되거나 데이터의 개수가 너무 적어질 가능성이 있음

```python
ebola_dropna = ebola.dropna()

print(ebola.shape)
print(ebola_dropna.shape)
```

### 누락값 포함된 데이터 계산하기
- 누락값이 하나라도 포함되어 있다면 계산 결과는 누락값
- skipna=True 를 통해 누락값 무시 후 계산 

```python
ebola.Cases_Guinea.sum(skipna=True)
```

```python
ebola.Cases_Guinea.sum(skipna=False)
```
