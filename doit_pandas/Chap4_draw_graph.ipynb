### Î∞ïÏä§Í∑∏ÎûòÌîÑ

```python

box_plot = plt.figure()

axes1 = box_plot.add_subplot(1, 1, 1)
axes1.boxplot(
    [
        tips[tips['sex'] == 'Female']['tip']
        ,tips[tips['sex'] == 'Male']['tip']
    ],
    labels=['Female', 'Male']
)
axes1.set_title('Boxplot of Tiips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

### Îã§Î≥ÄÎüâ Îç∞Ïù¥ÌÑ∞Î°ú Îã§Î≥ÄÎüâ Í∑∏ÎûòÌîÑ - ÏÇ∞Ï†êÎèÑ
- ÏÉâÏÉÅ, ÌÅ¨Í∏∞ Îì±ÏùÑ Îã§Î•¥Í≤å ÌïòÏó¨ Íµ¨Î∂ÑÌï† Ïàò ÏûàÏùå

```python
#0Í≥º 1Ïùò Ïà´ÏûêÎ°úÎßå ÏßÄÏ†ïÌï† Ïàò ÏûàÏúºÎØÄÎ°ú Sex Ïª¨ÎüºÏùò Î¨∏ÏûêÏó¥ÏùÑ Ïà´ÏûêÎ°ú Î∞îÍøîÏ£ºÎäî ÏûëÏóÖ ÏßÑÌñâ
#Î©îÏÑúÎìú Î∞òÌôòÍ∞íÏùÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Ï∂îÍ∞Ä

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill']
    ,y=tips['tip']
    ,s=tips['size'] * 10    #Ï†êÏùò ÌÅ¨Í∏∞
    ,c=tips['sex_color']   #Ï†êÏùò ÏÉâÏÉÅ
    ,alpha=0.5               #Ï†êÏùò Ìà¨Î™ÖÎèÑ
)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# ÏÇ∞Ï†êÎèÑ Í∑∏ÎûòÌîÑÏùò Îã®Ï†ê Î≥¥ÏôÑ -> Ïú°Í∞Å Í∑∏ÎûòÌîÑ

hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

# 04-5 seaborn ÎùºÏù¥Î∏åÎü¨Î¶¨Î°ú Í∑∏ÎûòÌîÑ Ïä§ÌÉÄÏùº ÏÑ§Ï†ïÌïòÍ∏∞

```python
sns.set_style('whitegrid')

fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='tip', hue='sex', data=tips, split=True)
```

```python
#Î™®Îì† Ïä§ÌÉÄÏùº Î¶¨Ïä§Ìä∏Ïóê Îã¥ÏïÑÎëêÍ≥† ÌïòÎÇòÏî© Ï†ÅÏö©ÌïòÍ∏∞

fig = plt.figure()

seaborn_style = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_style):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='tip', data=tips, ax=ax)
        violin.set_title(style)
        
fig.tight_layout()
```
